```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as plt
```

```python
house_pricing = pd.read_csv('house_pricing.csv')
```

```python
house_pricing
```

```python
sns.relplot(x='LotArea', y='SalePrice', data=house_pricing)
```

```python
parameter = house_pricing[(house_pricing['LotArea']< 20000) & (house_pricing['SalePrice']< 200000) ]
sns.relplot(x='LotArea', y='SalePrice', data=parameter)
```

```python
sns.relplot(x='LotFrontage', y='LotArea', data=house_pricing)
```

```python
parameter = house_pricing[(house_pricing['LotArea']< 100000) & (house_pricing['LotFrontage']< 200) ]
sns.relplot(x='LotFrontage', y='LotArea', data=parameter)
```
